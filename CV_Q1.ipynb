```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#generating images and adding gaussian noise to each image
def generate_images(size, gray_level, std):
    return np.full((size, size), gray_level) + np.random.normal(0, std, (size, size))

def box_filter(images):
  images = np.array(images)
  blur_images = cv2.blur(images,(3,3))
  return blur_images

#displaying generated images
def display_images(images):
  fig = plt.figure(figsize=(15, 6))

  for i, image in enumerate(images):
    fig.add_subplot(2, 5, i+1)
    plt.imshow(image, cmap = 'gray')
    plt.axis('off')
    plt.title(i+1)

  plt.show()

def EST_NOISE(images):
  m = len(images)
  image_mean = sum(images) / m
  num = 0
  for img in images:
    num += (image_mean - img) ** 2

  image_std = (num  / (m - 1)) ** 0.5
  return image_mean, image_std

#images size and gray_Level
size = 256
gray_level = 128

#Gaussian parameters
std = 2.0
mean = 0

images = [] 
for i in range(10):
  images.append(generate_images(size, gray_level, std))

display_images(images)

image_before_filter= []
image_before_filter = EST_NOISE(images)

print('Images of the mean and standard deviation \n 1. Mean \t \t 2. Standard Deviation')
display_images(image_before_filter)
print('\t\t Mean Values \t\t')
print(image_before_filter[0])
print('\t\t Standard Deviation \t\t')
print(image_before_filter[1])


blur_images = box_filter(images)
display_images(blur_images)

image_after_filter = []
image_after_filter = EST_NOISE(blur_images)
print('Images of the mean and standard deviation \n 1. Mean \t \t 2. Standard Deviation')
display_images(image_after_filter)
print('\t\t Mean Values \t\t')
print(image_after_filter[0])
print('\t\t Standard Deviation \t\t')
print(image_after_filter[1])
```
