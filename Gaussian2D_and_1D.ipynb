```python
import numpy as np
import matplotlib.pyplot as plt

def Gaussian_2D(sigma):
    size = int(6*sigma + 1)
    x, y = np.meshgrid(np.arange(-size//2 + 1, size//2 + 1), np.arange(-size//2 + 1, size//2 + 1))
    g = np.exp(-((x**2 + y**2)/(2.0*sigma**2))) / (2*np.pi*sigma**2)
    return g / g.sum(), x, y

def Gaussian_1D(sigma):
    size = int(6*sigma + 1)
    x = np.arange(-size//2 + 1, size//2 + 1)
    g = np.exp(-((x**2)/(2.0*sigma**2))) / (np.sqrt(2*np.pi)*sigma)
    return g / g.sum(), x

sigma = 1.4
g2d, x1, y1 = Gaussian_2D(sigma)
g1dx, y = Gaussian_1D(sigma)
g1dy, x = Gaussian_1D(sigma)

# Plot the 1D Gaussian filter masks
plt.plot(g1dx, y, 'r', label='1D Gaussian along x')
plt.legend()
plt.title("1D Gaussian Filter Masks")
plt.show()

plt.plot(x, g1dy, 'b', label='1D Gaussian along y')
plt.legend()
plt.title("1D Gaussian Filter Masks")
plt.show()

# Plot the 2D Gaussian filter mask
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x1, y1, g2d, 50, cmap='bone')
plt.title("2D Gaussian Filter Masks")
ax.view_init(35, 35)
```
